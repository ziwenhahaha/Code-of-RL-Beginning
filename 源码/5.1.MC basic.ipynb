```python
gamma = 0.9   #æŠ˜æ‰£å› å­ï¼Œè¶Šæ¥è¿‘0è¶Šè¿‘è§†

rows = 5      #è®°å¾—è¡Œæ•°å’Œåˆ—æ•°è¿™é‡Œè¦åŒæ­¥æ”¹
columns = 5
# gridworld = GridWorld_v2.GridWorld_v2(rows=rows, columns=columns, forbiddenAreaNums=8, targetNums=2, seed = 52,forbiddenAreaScore=-10)
# gridworld = GridWorld_v2.GridWorld_v2(desc = [".#",".T"])             #èµµè€å¸ˆ4-1çš„ä¾‹å­
# gridworld = GridWorld_v2.GridWorld_v2(desc = ["##.T","...#","...."])  #éšä¾¿å¼„çš„ä¾‹å­
gridworld = GridWorld_v2.GridWorld_v2(forbiddenAreaScore=-10, score=1,desc = [".....",".##..","..#..",".#T#.",".#..."]) 
gridworld.show()

trajectorySteps = 100

value = np.zeros(rows*columns)       #åˆå§‹åŒ–å¯ä»¥ä»»æ„ï¼Œä¹Ÿå¯ä»¥å…¨0
qtable = np.zeros((rows*columns,5))  #åˆå§‹åŒ–ï¼Œè¿™é‡Œä¸»è¦æ˜¯åˆå§‹åŒ–ç»´æ•°ï¼Œé‡Œé¢çš„å†…å®¹ä¼šè¢«è¦†ç›–æ‰€ä»¥æ— æ‰€è°“

# np.random.seed(50)
policy = np.eye(5)[np.random.randint(0,5,size=(rows*columns))] 
print(gridworld.showPolicy)
gridworld.showPolicy(policy)
# gridworld.showPolicy(policy)
```
